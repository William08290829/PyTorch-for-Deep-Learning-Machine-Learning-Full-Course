## Tensor datatypes
### ⚠️ PyTorch 常見錯誤：三大張量問題

> 在使用 PyTorch 和進行深度學習時，最常遇到的錯誤之一，就是「張量不對」，而這些錯誤大致可分為三種情況：

---

#### 1. ❌ Tensors not right datatype（資料型別錯誤）

- 張量的 `dtype` 不對，例如需要 `float32`，但你傳入的是 `int64`。
- 常見錯誤：使用 `CrossEntropyLoss()` 損失函數時，預測值應為 `float32`，但標籤 (label) 需為 `int64`。
- ✅ **解法**：
  ```python
  labels = labels.to(torch.int64)
  inputs = inputs.to(torch.float32)
  ```

#### 2. ❌ Tensors not right shape（張量形狀不對）
- 模型或運算需要特定形狀的張量，但你給的 shape 不正確。
- 常見錯誤訊息:

  ```
  RuntimeError: mat1 and mat2 shapes cannot be multiplied
  ```

- ✅ 解法：檢查並使用 .shape 或 .view()、.reshape()、.permute() 進行調整：

  ```python
  print(tensor.shape)
  tensor = tensor.view(batch_size, -1)  # 攤平成2D輸入

  ```
#### 3. ❌ Tensors not on the right device（裝置錯誤）

- 模型在 GPU 上 (cuda)，但資料在 CPU 上，或反之。

- 常見錯誤訊息：

  ```
  RuntimeError: Expected all tensors to be on the same device, but found at least two devices
  ```
- ✅ 解法：統一將張量與模型放在同一個裝置：

  ```python
  device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
inputs = inputs.to(device)

  ```

## One of the most common errors in deep learning (shape errors)

````python
# Shapes need to be in the right way
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]], dtype=torch.float32)

tensor_B = torch.tensor([[7, 10],
                         [8, 11],
                         [9, 12]], dtype=torch.float32)

# torch.matmul(tensor_A, tensor_B) # (this will error)
````

````python
# View tensor_A and tensor_B.T
print(tensor_A)
print(tensor_B.T)
````

````python
# The operation works when tensor_B is transposed
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}\n")
print(f"New shapes: tensor_A = {tensor_A.shape} (same as above), tensor_B.T = {tensor_B.T.shape}\n")
print(f"Multiplying: {tensor_A.shape} * {tensor_B.T.shape} <- inner dimensions match\n")
print("Output:\n")
output = torch.matmul(tensor_A, tensor_B.T)
print(output)
print(f"\nOutput shape: {output.shape}")
````

## Reshaping, stacking, squeezing and unsqueezing

````python
# Create a tensor
import torch
x = torch.arange(1., 8.)
x, x.shape
````

### Reshaping

````python
# Add an extra dimension
x_reshaped = x.reshape(1, 7)
x_reshaped, x_reshaped.shape
````

````python
# Change view (keeps same data as original but changes view)
# See more: https://stackoverflow.com/a/54507446/7900723
z = x.view(1, 7)
z, z.shape
````

````python
x[0] = 0
z
````

````python
x[0] = 1
# Changing z changes x
z[:, 0] = 5 # 把張量 z 中 第 0 欄（第一欄） 的所有列（row），都改成數值 5
# ↑   ↑
# |   └→ 第 0 欄
# └→ 所有列（用 : 表示）
z, x
````

### Stacking

````python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=0) # try changing dim to dim=1 and see what happens
x_stacked
````

````python
x_stacked2 = torch.stack([x, x, x, x], dim=1) # try changing dim to dim=1 and see what happens
x_stacked2
````

### Squeezing


> 把 shape 裡面是 1 的維度移除掉（例如 [1, 3, 1, 5] → [3, 5]），讓資料更精簡、便於後續處理。



````python
print(f"Previous tensor: {x_reshaped}")
print(f"Previous shape: {x_reshaped.shape}")

# Remove extra dimension from x_reshaped
x_squeezed = x_reshaped.squeeze()
print(f"\nNew tensor: {x_squeezed}")
print(f"New shape: {x_squeezed.shape}")
````

### Unsqueezing


````python
print(f"Previous tensor: {x_squeezed}")
print(f"Previous shape: {x_squeezed.shape}")

## Add an extra dimension with unsqueeze
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f"\nNew tensor: {x_unsqueezed}")
print(f"New shape: {x_unsqueezed.shape}")
````

### Permute


> permute(*dims)：傳入一個新的維度順序，把 tensor 的 shape 重排。



````python
# Create tensor with specific shape
x_original = torch.rand(size=(224, 224, 3))

# Permute the original tensor to rearrange the axis order
x_permuted = x_original.permute(2, 0, 1) # shifts axis 0->1, 1->2, 2->0

print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}")
````

````python
test1 = torch.rand(size=(2, 3, 4))
test1_permuted = test1.permute(2, 0, 1)
test1, test1_permuted
````

## Reproducibility (trying to take the random out of random)


Reproducibility（再現性 或 可重現性）是機器學習與科學研究中的一個重要概念，意思是：
> 🔁「在相同的條件下，別人（或你自己）能夠得到和你一樣的實驗結果。」



````python
import torch

# Create two random tensors
random_tensor_A = torch.rand(3, 4)
random_tensor_B = torch.rand(3, 4)

print(f"Tensor A:\n{random_tensor_A}\n")
print(f"Tensor B:\n{random_tensor_B}\n")
print(f"Does Tensor A equal Tensor B? (anywhere)")
random_tensor_A == random_tensor_B
````

### Set the random seed


> 每次你要「產生同樣的亂數結果」，就必須 重新把亂數種子設回去原本的值，因為亂數產生器的狀態在用過之後會改變



````python
import torch

# Set the random seed
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3, 4)
print(f"Tensor C:\n{random_tensor_C}\n")

# Have to reset the seed every time a new rand() is called
# Without this, tensor_D would be different to tensor_C
# torch.random.manual_seed(seed=RANDOM_SEED) # try commenting this line out and seeing what happens
random_tensor_D = torch.rand(3, 4)

print(f"Tensor D:\n{random_tensor_D}\n")
print(f"Does Tensor C equal Tensor D? (anywhere)")
random_tensor_C == random_tensor_D
````
